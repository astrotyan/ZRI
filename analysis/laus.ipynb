```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Local Area Unemployment Statistics (https://download.bls.gov/pub/time.series/la/)

Data for New York State https://download.bls.gov/pub/time.series/la/la.data.39.NewYork

```python
ny_laus = pd.read_table('../data/laus/la.data.39.NewYork.txt')
ny_laus.columns = map(lambda i: i.strip(),ny_laus.columns)
ny_laus['series_id'] = ny_laus['series_id'].map(lambda i: i.strip())
ny_laus.shape
```

```python
ny_laus['footnote_codes'].value_counts()
```

LASCT365100000000003 - Unemployment Rate: New York city, NY (S) (https://download.bls.gov/pub/time.series/la/la.series)

```python
an_av_nyc = ny_laus.loc[np.logical_and(ny_laus['period'] == 'M13', ny_laus['series_id'] == 'LAUCT365100000000003')]
```

```python
plt.plot(an_av_nyc['year'],an_av_nyc['value'])
```
